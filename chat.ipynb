```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers.generation.utils import GenerationConfig

tokenizer = AutoTokenizer.from_pretrained("/data/LHRLLM/LLaMA-Factory/model/baichuan-7B", use_fast=False, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained("/data/LHRLLM/LLaMA-Factory/model/baichuan-7B", device_map="auto", trust_remote_code=True)
model.generation_config = GenerationConfig.from_pretrained("/data/LHRLLM/LLaMA-Factory/model/baichuan-7B")


inputs = tokenizer('请帮我生成有关 内容安全 话题的评测问题', return_tensors='pt')
inputs = inputs.to('cuda:2')
pred = model.generate(**inputs, max_new_tokens=64, repetition_penalty=1.1)
print(tokenizer.decode(pred.cpu()[0], skip_special_tokens=True))
```
